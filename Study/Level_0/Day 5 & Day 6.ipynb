```python
# 코드 처리하기

def solution(code): # code는 문자열
    ret = '' 
    mode = 0
    
    for idx in range(0, len(code)):
        if mode == 0 and code[idx] == '1':
           mode = 1
        elif mode == 1 and code[idx] == '1':
           mode = 0
        elif mode == 0 and idx % 2 == 0:
           ret += code[idx]
        elif mode == 1 and idx % 2 ==1:
           ret += code[idx]
        else:
           continue
  
    if ret == '':
       return "EMPTY"
    else:
       return ret

# 색다른 풀이
# def solution(code):
#     return "".join(code.split("1"))[::2] or "EMPTY"
```

```python
# 이어 붙인 수

def solution(num_list):
    odd_str, even_str = '', ''
    
    for i in num_list:
        if i % 2 == 1:
           odd_str += f'{i}'
        else:
           even_str += f'{i}'  
    return int(odd_str) + int(even_str) 
```

```python
# 수 조작하기 1

def solution(n, control):
    for i in control:
       if i == 'w':
          n += 1
       elif i == 's':
          n -= 1
       elif i == 'd':
          n += 10
       elif i == 'a':
          n -= 10
       else:
          print('wrong input')
   
    return n
  
# 색다른 풀이
# def solution(n, control):
#     key = dict(zip(['w','s','d','a'], [1,-1,10,-10]))
#     return n + sum([key[c] for c in control])
# 사전으로 풀 수 있다는 포인트

# def solution(n, control):
#     answer = n
#     c = { 'w':1, 's':-1, 'd':10, 'a':-10}
#     for i in control:
#         answer += c[i]
#     return answer
```

```python
# 수 조작하기 2

def solution(numLog):
    answer = ''
    c = { 1:'w', -1:'s', 10:'d', -10:'a'}
    for i in range(1, len(numLog)):
        differ = numLog[i] - numLog[i-1]
        answer += c[differ]
    return answer
```

```python
# 수열과 구간 쿼리 2
def solution(arr, queries):
    answer = []
    for a,b,c in queries: # queries의 원소인 각 query의 구성원소를 지정
        temp = []
        for i in sorted(arr[a:b+1]):
            if i > c:
               temp.append(i)
        if temp == []:
            answer.append(-1)
        else:
            answer.append(temp[0])
    return answer        
```
